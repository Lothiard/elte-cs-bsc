Line plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]

plt.plot(x,y,linewidth=2,marker="x")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line plot")
plt.show()
```

Scatter plot

```python
x = [1,2,3,4,5]
y = [5,3,4,2,7]

plt.scatter(x,y, s=80, alpha=0.7)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
categories = ["a","b","c","d"]
values =[10,14,8,12]

plt.bar(categories,values)
plt.show()
```

```python
import numpy as np
```

```python
data = np.random.randn(500)

plt.hist(data, bins=100, alpha=0.7)
plt.show()
```

```python
data = [np.random.randn(100), np.random.randn(100) + 2]

plt.boxplot(data, labels=['Group 1', 'Group 2'], showmeans=True)
plt.show()
```

```python
sizes = [40,30,20,10]
labels = ["Apple", "Banana", "Cherry", "Date"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.show()
```

```python
matrix = np.random.rand(5,5)

plt.imshow(matrix, cmap='viridis')
plt.colorbar()
plt.show()
```
